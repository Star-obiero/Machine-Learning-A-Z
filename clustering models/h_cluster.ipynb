Hierachical Clustering Model

Import libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

Load data set

```python
df=pd.read_csv(".\\Mall_Customers.csv")
x=df.iloc[:,[3,4]].values
```

Using the dendogram to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendogram')
plt.xlabel('customers')
plt.ylabel('Euclidean distances')
plt.show()
```

Training the Hierachical clustering model

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,metric='euclidean',linkage='ward')
y_hc=hc.fit_predict(x)
```

```python
print(y_hc)
```

Visualizing the clusters

```python
plt.scatter(x[y_hc==0,0],x[y_hc==0,1],s=100,color='red',label='cluster 1')
plt.scatter(x[y_hc==2,0],x[y_hc==2,1],s=100,color='blue',label='cluster 2')
plt.scatter(x[y_hc==3,0],x[y_hc==3,1],s=100,color='green',label='cluster 3')
plt.scatter(x[y_hc==4,0],x[y_hc==4,1],s=100,color='magenta',label='cluster 4')
plt.scatter(x[y_hc==5,0],x[y_hc==5,1],s=100,color='cyan',label='cluster 5')
plt.title("Cluster of customers")
plt.xlabel("Anuall income(ksh)")
plt.ylabel('Spending score (1-100)')
plt.legend()
plt.show()
```
