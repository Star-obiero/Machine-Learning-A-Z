# Knowing when to use multiple linear regression

Import libries

```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

Loading the dataset

```python
df=pd.read_csv(".\\Data.csv")
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
y=y.reshape(len(y),1)
y
```

Splitting the datase into training and testing set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
```

Training the multile liner regression on a trainig set

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

Predicting the test set result

```python
y_pred=regressor.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

Evaluating the model performance


```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

# Polynomial Regression

 Importing libraries

```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

Loading a dataset

```python
df=pd.read_csv(".\\Data.csv")
x2=df.iloc[:,:-1].values
y2=df.iloc[:,-1].values
```

Splitting the data into the traing set and test set

```python
from sklearn.model_selection import train_test_split
x_train2,x_test2,y_train2,y_test2=train_test_split(x2,y2,test_size=0.2,random_state=40)
```

Training the Polynomial regression model in the training set

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_reg=PolynomialFeatures(degree=4)
x_poly=poly_reg.fit_transform(x_train2)
regressor=LinearRegression()
regressor.fit(x_poly,y_train2)
```

Predicting the test results

```python
y_pred=regressor.predict(poly_reg.transform(x_test2))
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test2),1)),1))
```

Evaluating the model performance

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
